```python
# Executive Summary: Customer Segmentation
```

```python
## Objective
Segment customers based on purchasing behavior to enable targeted marketing and retention strategies.

## Methodology
- Cleaned and preprocessed retail transaction data.
- Performed RFM (Recency, Frequency, Monetary) analysis.
- Applied KMeans clustering to group customers into 4 segments.
```

```python
## Key Insights
- Identified high-value, loyal, and at-risk customer segments.
- High-value customers represent a significant portion of revenue.
- At-risk customers have not purchased in the last 6 months.
- Most customers fall into 2 main segments, with clear differences in recency and monetary value.
```

```python
## 5. Business Recommendations

- High-value customers: Target with loyalty programs.
- At-risk customers: Send re-engagement offers.
- New/low-value customers: Encourage with onboarding and incentives.
```
